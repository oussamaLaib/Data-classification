## Data preparation

Since I'm using Artificial Neural Networks (RNN) for the classification modeling, in order to achieve a better calculation of network weights, using an appropriate approach to normalize the provided data is critical before feeding it to RNN models. As all the models are designed to use the Sigmoid as activation function, the inputs data are normalized to an interval between [0,1].

In the case of adopting and developing RNN, three separate datasets should be prepared.
    - One for test, to assess the model's performance (20%).
    - the rest is split between training and validation (80% and 20% respectively), where the training dataset is used to adjust the model's parameters, and the validation set is used to avoid falling in the overfitting phenomenon.

As we can see in the above plot, the dataset is composed of more than 1000 of the first class and less than 200 of the second class, which means that the dataset is highly imbalanced. To solve the data imbalanced issue, it is required to proportionally split the data between training and test set, this could be achieved by using the 'StratifiedShuffleSplit' found in 'sklearn'. 

```python
scaler= MinMaxScaler(feature_range=(0,1))
Features = pd.DataFrame(scaler.fit_transform(Features))
# X_test = pd.DataFrame(scaler.fit_transform(X_test))
# X_val = pd.DataFrame(scaler.fit_transform(X_val))


ss = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=2)
for training_index, test_index in ss.split(Features, Classes):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_training, X_test = Features.loc[training_index,], Features.loc[test_index,]
    y_training, y_test = Classes.loc[training_index,], Classes.loc[test_index,]


sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=2)
for train_index, val_index in sss.split(X_training, y_training):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_val = Features.loc[train_index,], Features.loc[val_index,]
    y_train, y_val = Classes.loc[train_index,], Classes.loc[val_index,]
```

### Classification modeling using LSTM RNN


Applying RNN requires adjusting different parameters, including: number of hidden layers, number of neurons within each layer and learning rate value. To find the best combination of these parameters, Random Search is used. 

```python
def LSTM_model( SIZE_HIDDEN_LAYERS = [ 20, 10, 5], LEARNING_RATE = .05):
    #####################
    # Pre training
    #####################

    INPUTS_TRAIN = X_train.copy()
    OUTPUTS_TRAIN = y_train.copy()
    
    INPUTS_VAL = X_val.copy()
    OUTPUTS_VAL = y_val.copy()

    model = Sequential()

    if len(SIZE_HIDDEN_LAYERS) > 1:
        model.add( LSTM( units = SIZE_HIDDEN_LAYERS[0], return_sequences = True, activation='sigmoid',
                    input_shape = ( 1, INPUTS_TRAIN.shape[1])))  
        model.add( LSTM( units = SIZE_HIDDEN_LAYERS[1], activation = 'sigmoid', return_sequences=False))
        
    else:
        model.add( LSTM( units = SIZE_HIDDEN_LAYERS[0], return_sequences = False, activation='sigmoid',
                    input_shape = ( 1, INPUTS_TRAIN.shape[1])))  
    
    model.add( Dense( 1, init ='uniform',activation = 'sigmoid'))
    optimizer=Adamax(lr = LEARNING_RATE)
    model.compile(loss = 'binary_crossentropy', optimizer = optimizer, metrics = ['acc','mse'])

    model.summary()
    
    
    
    inputs = np.array(INPUTS_TRAIN).reshape(-1, 1, INPUTS_TRAIN.shape[1])
    outputs = np.array(OUTPUTS_TRAIN).reshape(OUTPUTS_TRAIN.shape[0],-1)
    
    inputs_val = np.array(INPUTS_VAL).reshape(-1, 1, INPUTS_VAL.shape[1])
    outputs_val = np.array(OUTPUTS_VAL).reshape(OUTPUTS_VAL.shape[0],-1)
    
    
    counter = 0
    Epoch = 200
    for i in range(0,Epoch):
        print('epoch:   %d .' %counter)
        ErrorHistory = model.fit(x = inputs, y = outputs, epochs = 1, batch_size = 12
                                   , validation_data = (inputs_val, outputs_val), verbose = 2)


    file_name = 'Experiments/Models/LSTMs/lstm_h_{hidden_l_sizes}_l_{learning_rate}.h5'.format(
        hidden_l_sizes = '_'.join([str(s) for s in SIZE_HIDDEN_LAYERS]),
        learning_rate = str(LEARNING_RATE))
    model.save(file_name)  # creates a HDF5 file 'my_model.h5'
    c = pd.DataFrame(ErrorHistory.history)
    hfile_name = 'Experiments/Models/LSTMs/lstm_h_{hidden_l_sizes}_l_{learning_rate}.csv'.format(
        hidden_l_sizes = '_'.join([str(s) for s in SIZE_HIDDEN_LAYERS]),
        learning_rate = str(LEARNING_RATE))
    c.to_csv(hfile_name)
    return model
```

During the Random search process, the model of each experiment is saved to be chosen later depending on its performance (considering the mean absolute error and the classification accuracy) 

```python
def train(*args, **kwargs):
    global experiment_index
    global experiments

    global X_train
    global X_test
    global X_val
    
    global y_train
    global y_test
    global y_val
    
    size_hidden_layers=[]
    size_hidden_layers.append(int(kwargs.get('size_hidden_layers_1')))
    if kwargs.get('size_hidden_layers_2', None):
        size_hidden_layers.append(int(kwargs.get('size_hidden_layers_2')))

    
    model = LSTM_model(SIZE_HIDDEN_LAYERS = size_hidden_layers,
                 LEARNING_RATE = kwargs.get('learning_rate'))

    # generate predictions for training
    
    trainPredict = model.predict_classes(np.array(X_train).reshape(-1, 1, X_train.shape[1]))
    testPredict = model.predict_classes(np.array(X_test).reshape(-1, 1, X_test.shape[1]))
    validationPredict = model.predict_classes(np.array(X_val).reshape(-1, 1, X_val.shape[1]))

    y_train = np.array( y_train)
    y_test = np.array( y_test)
    y_val = np.array( y_val)

    ACC_train = accuracy_score(y_train, trainPredict) 
    ACC_test = accuracy_score(y_test, testPredict) 
    ACC_validation = accuracy_score(y_val, validationPredict) 


    mae_train = mean_absolute_error(y_train, trainPredict)
    mae_validation = mean_absolute_error(y_val, validationPredict)
    mae_test = mean_absolute_error(y_test, testPredict)

    b = pd.DataFrame({
        'size': len(size_hidden_layers),
        'learning_rate': kwargs.get('learning_rate'),
        'hidden_l_sizes': '_'.join([str(s) for s in size_hidden_layers]),
        'Train_ACC': ACC_train,
        'Train_MAE': mae_train,
        'Validation_ACC': ACC_validation,
        'Validation_MAE': mae_validation,
        'Test_ACC': ACC_test,
        'Test_MAE': mae_test}, index = [experiment_index])
    experiments = pd.concat([experiments, b])
    print('Train ACC = %s %%' % ACC_train)
    print('Valdiation ACC = %s %%' % ACC_validation)
    print('Test ACC = %s %%' % ACC_test)
    experiment_index = experiment_index + 1
    return ACC_validation
```

# Random Search experiments

Thirty experiments is performed for both models with one hidden layer and two hidden layers. The number of hidden layers varies from 30 to 80 neurons in the first hidden layer and from 10 to 30 in the second hidden layer. The learning rate value varies from 0.001 to 0.01.

The number of epochs for the weights adjustment is 200 with a mini-batch size of 12.

The final result of the experiments is saved as a CSV file name 'lstm_experiments' in the Experiments folder, which summarises the performence of each model alongside with its hyper-parameters.

```python
NUM_EVALS=30


experiment_index = 1
experiments = pd.DataFrame()
solver = optunity.make_solver('random search', num_evals = NUM_EVALS,
                              learning_rate = [0.001, 0.01],
                              size_hidden_layers_1 = [30, 80])
result, details = solver.minimize(train)

total_experiments = experiments


experiment_index = 1
experiments = pd.DataFrame()
solver = optunity.make_solver('random search', num_evals = NUM_EVALS,
                              learning_rate = [0.001, 0.10],
                              size_hidden_layers_1 = [30, 80],
                              size_hidden_layers_2 = [10, 30])
result, details = solver.minimize(train)

frames = [total_experiments, experiments]
total_experiments = pd.concat(frames)


total_experiments.to_csv('Experiments/lstm_experiments.csv')
```

# Final Results

Looking at the performance of each model in terms of classification accuracy using the test set, the highest recorded accuracy is performed by the 17th model with 1 hidden layer.
The accuracy on test, training and validation phases is: 0.912%, 0.975% and 0.937% respectively.
