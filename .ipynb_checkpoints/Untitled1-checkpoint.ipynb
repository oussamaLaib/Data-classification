```python
def import_data():
    dataset = pd.read_csv('./Data/Data.csv',engine='python')
    return dataset

def mean_absolute_percentage_error(y_true, y_pred): 

    ## Note: does not handle mix 1d representation
    #if _is_1d(y_true): 
    #    y_true, y_pred = _check_1d_array(y_true, y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def mean_absolute_error(y_true, y_pred): 

    ## Note: does not handle mix 1d representation
    #if _is_1d(y_true): 
    #    y_true, y_pred = _check_1d_array(y_true, y_pred)
    return np.mean(np.abs((y_true - y_pred) )) 

def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())

```

```python
from IPython.display import clear_output

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import learning_curve

import keras
from keras.models import load_model
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.layers.core import  Activation, Dropout  
from keras.optimizers import RMSprop
from keras.layers import Embedding, RepeatVector
from keras.layers import LSTM
from keras.layers.recurrent import GRU
from keras.optimizers import SGD
from sklearn.preprocessing import MinMaxScaler

from keras.layers.convolutional import Conv3D
from keras.layers.convolutional_recurrent import ConvLSTM2D
from keras.layers.normalization import BatchNormalization


from keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Dropout, Flatten, ConvLSTM2D
from keras import optimizers
from keras.layers import TimeDistributed





import datetime
```

```python
def create_Train_LSTM_CNN_Model(inputs, outputs, inputs_val, outputs_val, HiddenNeurons1 = None, HiddenNeurons2 = None
                              , LearningRate = None, Epoch = None, Optimizer = None):
    
    if HiddenNeurons1 is None: HiddenNeurons1 = 10
    if HiddenNeurons2 is None: HiddenNeurons2 = 5
    if LearningRate is None: LearningRate = 0.001
    if Epoch is None: Epoch = 500
    if Optimizer is None: Opitimizer = 'Adam'
        
    RCNNmodel = Sequential()

    RCNNmodel.add(ConvLSTM2D(filters=10, kernel_size=(1, 8),
    input_shape=(1, 1, inputs.shape[3], 1),
    padding='same', return_sequences=True))
    RCNNmodel.add(BatchNormalization())
    RCNNmodel.add(ConvLSTM2D(filters=4, kernel_size=(1, 3),padding='same', return_sequences=True))
    RCNNmodel.add(Flatten())
    RCNNmodel.add(Dense(1, activation='sigmoid'))
    RCNNmodel.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    RCNNmodel.summary()




    tempRCNNmodels= np.empty((Epoch),dtype=object)
    tempErrorHistory= np.empty([Epoch,4], dtype=float)


    
#     inputs= np.array(inputs).reshape(-1, 1, inputs.shape[1])
    outputs= np.array(outputs).reshape(inputs.shape[0],-1)
    
#     inputs_val= np.array(inputs_val).reshape(-1, 1, inputs_val.shape[1])
    outputs_val= np.array(outputs_val).reshape(outputs_val.shape[0],-1)
    
    counter=0
    for i in range(0,Epoch):
        print('epoch:   %d .' %counter)
        ErrorHistory = RCNNmodel.fit(x = inputs, y = outputs, epochs = 1, batch_size = 12
                                   , validation_data = (inputs_val, outputs_val), verbose = 2)
#         ErrorHistory= mlpmodel.fit(x= inputs.reshape(-1, 1, inputs.shape[1]), y = outputs, epochs=1,batch_size=12
#                            ,validation_split= 0.2,verbose= 2)


        
        tempRCNNmodels[i]= RCNNmodel
        RCNNmodel.save('Temps/RCNNs/rcnn_%d'%i)
        tempErrorHistory[i,0] = ErrorHistory.history['val_loss'][0]
        tempErrorHistory[i,1] = ErrorHistory.history['loss'][0]
        tempErrorHistory[i,2] = ErrorHistory.history['val_acc'][0]
        tempErrorHistory[i,3] = ErrorHistory.history['acc'][0]
        counter = counter +1
    
    print('==============================================================================')
    print('\t \t LSTM model is ready for forecasting')
    print('==============================================================================')
    
    
    return tempRCNNmodels, tempErrorHistory, RCNNmodel
```

```python
scaler= MinMaxScaler(feature_range=(0,1))
Features = pd.DataFrame(scaler.fit_transform(Features))
# X_test = pd.DataFrame(scaler.fit_transform(X_test))
# X_val = pd.DataFrame(scaler.fit_transform(X_val))


ss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=2)
for training_index, test_index in ss.split(Features, Classes):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_training, X_test = Features.loc[training_index,], Features.loc[test_index,]
    y_training, y_test = Classes.loc[training_index,], Classes.loc[test_index,]


sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=2)
for train_index, val_index in sss.split(X_training, y_training):
    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_val = Features.loc[train_index,], Features.loc[val_index,]
    y_train, y_val = Classes.loc[train_index,], Classes.loc[val_index,]

X_train= np.array(X_train).reshape(X_train.shape[0],1, 1, X_train.shape[1],1)
X_test= np.array(X_test).reshape(X_test.shape[0],1, 1, X_test.shape[1],1)
X_val= np.array(X_val).reshape(X_val.shape[0],1, 1, X_val.shape[1],1)


# X_train = pd.DataFrame(X_train.reshape(-1,1, X_train.shape[1]))
# X_test = pd.DataFrame(X_test.reshape(-1,1, X_test.shape[1]))
# X_val = pd.DataFrame(X_val.reshape(-1,1, X_val.shape[1]))
Epochs= 200
HiddenNeurons1 = 80
HiddenNeurons2 = 35
LearningRate = 0.001

tempMLPmodels, tempErrorHistory, RCNN_Model = create_Train_LSTM_CNN_Model(X_train, y_train, X_val, y_val, HiddenNeurons1 = HiddenNeurons1,HiddenNeurons2 = HiddenNeurons2
                                      , Epoch = Epochs, LearningRate = LearningRate, Optimizer = 'Adamax')


```
